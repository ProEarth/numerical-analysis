```python
import numpy as np
import matplotlib.pyplot as plt


p_cruise = 0.9
n_passengers = 3
p_pass = 0.8


num_simulations = 40000


X = np.random.binomial(1, p_cruise, num_simulations)


Y = np.random.binomial(n_passengers, p_pass, num_simulations)


Z = X * Y


unique, counts = np.unique(Z, return_counts=True)
frequencies = counts / num_simulations * 100


plt.bar(unique, frequencies, width=0.6, alpha=0.7, color='b', edgecolor='black')


plt.xlabel("Total # of Passengers Who Show Up for Flight")
plt.ylabel("# of Simulations (%)")
plt.title("Simulation of XYZ Airlines Passenger Arrivals")


for i in range(len(unique)):
    plt.text(unique[i], frequencies[i] + 0.5, f"{frequencies[i]:.1f}%", ha='center', fontsize=10)

plt.show()
```
